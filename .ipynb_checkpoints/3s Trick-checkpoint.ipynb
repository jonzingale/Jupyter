## The 3s Trick for English Majors.

Nick,

I will here attempt a proof of the threes trick for English majors. As is readily noticed, whenever a numbers digits are divisible by three the number itself is divisible by three. By divisible we mean that whenever a number `n` is divided by another `d` there is a remainder of zero. But what is a remainder?<p>

Definition. (Remainder)<br>
A natural number `r` is a remainder of `d` dividing `n` whenever one can find an integer `q` such that `n = dq + r`. The remainder is the smallest such `r`. For instance, if `d=5` and `n=23` we can find an integer `q=3` giving `23 = 5(4) + r`. Here the remainder is `r=3`. For notational simplicity in what follows, let $[n]_d$ denote the remainder when `n` is divided by `d`. Note that when `d` is a factor of `n` then $[n]_d$ = 0.<p>

The entire game of proving the 3s trick will rely on understanding how the remainder of a division problem is effected by adding and multiplying numbers. If a number `n` is divisible by three and adding to `n` in clever ways and multiplying by `n` in clever ways leave the remainder unchanged, whatever number is finally arrived at will also be divisible by three. The following lemmas with help to establish what is meant by clever.<br> 

Lemma 1.<br>
Remainders are 'preserved' by Addition, ie. Summing two numbers and then dividing by `d` gives the same remainder as dividing both numbers by `d`, summing the remainders and then again dividing by `d`. In the fancy short hand of the definition above: $[n+m]_d=[[n]_d+[m]_d]_d$<p>

proof:<br>
Let `n` and `m` be integers which can be written in the forms `n = dq + r` and `m = dp + s`, with d being the same in both expressions. Here `r` is the remainder when dividing `n` by `d` and `s` is the remainder when dividing `m` by `d`. It can be seen algebraically that:<p>

$n+m$
$(dq + r)+(dp+s)$ # by assumption<br>
$(dq + dp) + (r+s)$ # associating the parts with ds and then those without.<br>
$d(q+p) + (r+s)$<p>
`d` divides `d(q + p)` and so `r + s` determines the remainder of dividing `n + m` by `d`.<p>

Corollary.<br>
Two numbers divisible by `d` yield a sum divisible by `d`,<br>
$[n+m]_d = 0 \implies [[n]_d + [m]_d]_d = 0$<p>

Lemma 2.<br>
Remainders are preserved by multiplication,
$[nm]_d=[[n]_m[m]_d]_d$<p>

proof:<br>
Similar to the proof of Lemma 1, we can write integers<br>
`n` and `m` in the forms `n = dq + r` and `m = dp + s`.<br>
Algebraically it follows that:<p>

$n*m$<br>
$(dq+r)(dp+s)$<br>
$d^2pq+dqs+dpr+rs$<br>
$d(dpq+qs+pr)+rs$<p>
`d` divides $d(dpq+qs+pr)$ and so does not contribute to any change in the remainder.<br>
Thus, the remainder of dividing `n * m` by d is determined by `r*s`.<p>

Now we are ready to prove the main theorem.<p>

Theorem. (3s trick: If 3 divides the sum of ns digits, 3 divides n)<p>

proof:<br>
Let's begin by remembering that every natural number `n` can be written in what is called its m-ary representation, namely, as $n=\sum_{i=0}a_i*10^i=a_0+a_1+a_2+...+a_n$. For example, the number 1234 can be written as $1*10^0+2*10^1+3*10^2+4*10^3$. By the corollary to Lemma 1, if dividing a sum of a bunch of numbers by 3 has a remainder of zero then dividing a bunch of numbers by 3 and summing their remainders will give a remainder of zero (effectively), ie.<p>
$[\sum_{i=0}a_i]_3=[\sum_{i=0}{[a_i]}]_3$<p>
Dividing 10 by 3 gives a remainder of 1 and multiplication preserves remainders by Lemma 2. This means that multiplying a number by 10 or any powers of 10 will not change the remainder. This is because multiplying by 10 has the effect of multiplying the remainder by 1.<p>
Given that $\sum_{i=0}a_i=a_0+a_1+a_2+...+a_n$ multiplying any individual coefficients $a_i$ by a power of 10, $10^i$, will leave the remainder unchanged.<br>
Thus, $[\sum_{i=0}{a_i*10^i}]_3=0$, ie.<br>$\sum_{i=0}{a_i*10^i}$ is divisible by 3.<p>

Again in the fancy notation:<p>
$[\sum{a_i}]_d=0$ # by premise<br>
$\sum{[a_i]_d}=0$ # by Lemma 1<br>
$\sum{[a_i]_d*[1]_d}=0$<br>
$\sum{[a_i]_d*[10^i]_d}=0$ # by substitional equivalence<br>
$\sum{[a_i*10^i]_d}=0$ # by Lemma 2<br>
$[\sum{a_i*10^i}]_d=0$ # by Lemma 1<br>
3 divides $\sum{a_i*10^i}$ # by definition of remainders.<p>
